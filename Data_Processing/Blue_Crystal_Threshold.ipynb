```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['lines.markersize'] = 3  # Change this value as needed
```

```python
cal = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/cristal azul/with and without cristal.txt'

df_cal = pd.read_table(cal, decimal=',')

df_cal.columns = ['Threshold - No Scintillator', 'Frequency - No Scintillator', 'Threshold - With Scintillator', 'Frequency - With Scintillator']

# Processing for the radioactive samples

bAm1 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/cristal azul/Am 1.txt'
bAm2 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/cristal azul/Am 2.txt'
bAm3 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/cristal azul/Am 3.txt'

bCo1 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/cristal azul/Co 1.txt'
bCo2 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/cristal azul/Co 2.txt'
bCo3 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/cristal azul/Co 3.txt'

bCs1 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/cristal azul/Cs 1.txt'
bCs2 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/cristal azul/Cs 2.txt'
bCs3 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/cristal azul/Cs 3.txt'

df_bAm1 = pd.read_table(bAm1, decimal=',')
df_bAm2 = pd.read_table(bAm2, decimal=',')
df_bAm3 = pd.read_table(bAm3, decimal=',')

df_bCo1 = pd.read_table(bCo1, decimal=',')
df_bCo2 = pd.read_table(bCo2, decimal=',')
df_bCo3 = pd.read_table(bCo3, decimal=',')

df_bCs1 = pd.read_table(bCs1, decimal=',')
df_bCs2 = pd.read_table(bCs2, decimal=',')
df_bCs3 = pd.read_table(bCs3, decimal=',')

df_bAm = [df_bAm1, df_bAm2, df_bAm3]
df_bCo = [df_bCo1, df_bCo2, df_bCo3]
df_bCs = [df_bCs1, df_bCs2, df_bCs3]

for i in df_bAm:
    i.drop(i.columns[[0, 1, 2, 3]], axis=1, inplace=True)
    i.columns = ['Threshold', 'Frequency']

for i in df_bCo:
    i.drop(i.columns[[0, 1, 2, 3]], axis=1, inplace=True)
    i.columns = ['Threshold', 'Frequency']

for i in df_bCs:
    i.drop(i.columns[[0, 1, 2, 3]], axis=1, inplace=True)
    i.columns = ['Threshold', 'Frequency']
```

```python
# Checking if all the trials are good and generating the mean result

for i, df in enumerate(df_bAm, start=1):
    plt.semilogy(df['Threshold'], df['Frequency'], label=f'Trial {i}')

plt.title('Blue Am Trials')
plt.xlabel('Threshold / mV')
plt.ylabel('Frequency / kHz')

mean_df_bAm = pd.concat(df_bAm).groupby(level=0).mean()

plt.semilogy(mean_df_bAm['Threshold'], mean_df_bAm['Frequency'], label=f'Mean')

# Plotting the calibration measurements
n = 5
plt.semilogy(df_cal['Threshold - No Scintillator'], df_cal['Frequency - No Scintillator'], label='No Scintillator')
plt.semilogy(df_cal['Threshold - With Scintillator'][n:], df_cal['Frequency - With Scintillator'][n:], label='Only Scintillator')

plt.grid()
plt.legend()

plt.savefig('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Data_Processing/Blue Am.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Checking if all the trials are good and generating the mean result

for i, df in enumerate(df_bCo, start=1):
    plt.semilogy(df['Threshold'], df['Frequency'], label=f'Trial {i}')

plt.title('Blue Co Trials')
plt.xlabel('Threshold / mV')
plt.ylabel('Frequency / kHz')

# Getting and scatterting the mean measurements
mean_df_bCo = pd.concat(df_bCo).groupby(level=0).mean()

plt.semilogy(mean_df_bCo['Threshold'], mean_df_bCo['Frequency'], label=f'Mean')

# Plotting the calibration measurements
n = 5
plt.semilogy(df_cal['Threshold - No Scintillator'], df_cal['Frequency - No Scintillator'], label='No Scintillator')
plt.semilogy(df_cal['Threshold - With Scintillator'][n:], df_cal['Frequency - With Scintillator'][n:], label='Only Scintillator')

plt.grid()
plt.legend()

plt.savefig('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Data_Processing/Blue Co.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Checking if all the trials are good and generating the mean result

for i, df in enumerate(df_bCs, start=1):
    plt.semilogy(df['Threshold'], df['Frequency'], label=f'Trial {i}')

plt.title('Blue Cs Trials')
plt.xlabel('Threshold / mV')
plt.ylabel('Frequency / kHz')

# Getting and scatterting the mean measurements
mean_df_bCs = pd.concat(df_bCs).groupby(level=0).mean()

plt.semilogy(mean_df_bCs['Threshold'], mean_df_bCs['Frequency'], label=f'Mean')

# Plotting the calibration measurements
n = 5
plt.semilogy(df_cal['Threshold - No Scintillator'], df_cal['Frequency - No Scintillator'], label='No Scintillator')
plt.semilogy(df_cal['Threshold - With Scintillator'][n:], df_cal['Frequency - With Scintillator'][n:], label='Only Scintillator')

plt.grid()
plt.legend()

plt.savefig('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Data_Processing/Blue Cs.png', dpi=300, bbox_inches='tight')

plt.show()
```
