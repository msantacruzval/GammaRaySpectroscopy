```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
rAmG30 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal rojo (Csi)/Am G30.txt'
rAmG40 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal rojo (Csi)/Am G40.txt'
rAmG50 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal rojo (Csi)/Am G50.txt'

rCsG30 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal rojo (Csi)/Cs G30.txt'
rCsG40 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal rojo (Csi)/Cs G40.txt'
rCsG50 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal rojo (Csi)/Cs G50.txt'

rCoG30 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal rojo (Csi)/Co G30.txt'
rCoG40 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal rojo (Csi)/Co G40.txt'
rCoG50 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal rojo (Csi)/Co G50.txt'

df_rAmG30 = pd.read_table(rAmG30, decimal=',')
df_rAmG40 = pd.read_table(rAmG40, decimal=',')
df_rAmG50 = pd.read_table(rAmG50, decimal=',')

df_rCsG30 = pd.read_table(rCsG30, decimal=',')
df_rCsG40 = pd.read_table(rCsG40, decimal=',')
df_rCsG50 = pd.read_table(rCsG50, decimal=',')

df_rCoG30 = pd.read_table(rCoG30, decimal=',')
df_rCoG40 = pd.read_table(rCoG40, decimal=',')
df_rCoG50 = pd.read_table(rCoG50, decimal=',')

df_rAm = [df_rAmG30, df_rAmG40, df_rAmG50]
df_rCs = [df_rCsG30, df_rCsG40, df_rCsG50]
df_rCo = [df_rCoG30, df_rCoG40, df_rCoG50]

for i in df_rAm:
    i.columns = ['ADC', 'Counts']

for i in df_rCs:
    i.columns = ['ADC', 'Counts']

for i in df_rCo:
    i.columns = ['ADC', 'Counts']
```

```python
for i, df in enumerate(df_rAm, start=1):
    plt.hist(df['ADC'],
             bins=2000,
             weights=df['Counts'],
             alpha=0.7,
             log=False,
             label=f'Gain {10*i+20}')


plt.title('Red Crystal Am')
plt.legend()
plt.grid()

plt.savefig('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Data_Processing/Red Am Spectrum.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
for i, df in enumerate(df_rCs, start=1):
    plt.hist(df['ADC'],
             bins=2000,
             weights=df['Counts'],
             alpha=0.7,
             log=False,
             label=f'Gain {10*i+20}')


plt.title('Red Crystal Cs')
plt.legend()
plt.grid()

plt.savefig('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Data_Processing/Red Cs Spectrum.png', dpi=300, bbox_inches='tight')

plt.show()
```
