```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
nAmG30 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal blank/Am G30.txt'
nAmG40 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal blank/Am G40.txt'
nAmG50 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal blank/Am G50.txt'

nCsG30 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal blank/Cs G30.txt'
nCsG40 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal blank/Cs G40.txt'
nCsG50 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal blank/Cs G50.txt'

nCoG30 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal blank/Co G30.txt'
nCoG40 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal blank/Co G40.txt'
nCoG50 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal blank/Co G50.txt'

df_nAmG30 = pd.read_table(nAmG30, decimal=',')
df_nAmG40 = pd.read_table(nAmG40, decimal=',')
df_nAmG50 = pd.read_table(nAmG50, decimal=',')

df_nCsG30 = pd.read_table(nCsG30, decimal=',')
df_nCsG40 = pd.read_table(nCsG40, decimal=',')
df_nCsG50 = pd.read_table(nCsG50, decimal=',')

df_nCoG30 = pd.read_table(nCoG30, decimal=',')
df_nCoG40 = pd.read_table(nCoG40, decimal=',')
df_nCoG50 = pd.read_table(nCoG50, decimal=',')

df_nAm = [df_nAmG30, df_nAmG40, df_nAmG50]
df_nCs = [df_nCsG30, df_nCsG40, df_nCsG50]
df_nCo = [df_nCoG30, df_nCoG40, df_nCoG50]

for i in df_nAm:
    i.columns = ['ADC', 'Counts']

for i in df_nCs:
    i.columns = ['ADC', 'Counts']

for i in df_nCo:
    i.columns = ['ADC', 'Counts']
```

```python
for i, df in enumerate(df_nAm, start=1):
    plt.hist(df['ADC'],
             bins=500,
             weights=df['Counts'],
             alpha=0.7,
             log=True,
             label=f'Gain {10*i+20}')


plt.title('Blank Crystal Am')
plt.legend()
plt.grid()

plt.savefig('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Data_Processing/Blank Am Spectrum.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
for i, df in enumerate(df_nCs, start=1):
    plt.hist(df['ADC'],
             bins=500,
             weights=df['Counts'],
             alpha=0.7,
             log=True,
             label=f'Gain {10*i+20}')


plt.title('Blank Crystal Cs')
plt.legend()
plt.grid()

plt.savefig('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Data_Processing/Blank Cs Spectrum.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
for i, df in enumerate(df_nCo, start=1):
    plt.hist(df['ADC'],
             bins=250,
             weights=df['Counts'],
             alpha=0.7,
             log=True,
             label=f'Gain {10*i+20}')


plt.title('Blank Crystal Co')
plt.legend()
plt.grid()

plt.savefig('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Data_Processing/Blank Co Spectrum.png', dpi=300, bbox_inches='tight')

plt.show()
```
