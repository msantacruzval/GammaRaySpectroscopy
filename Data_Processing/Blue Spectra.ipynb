```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
bAmG30 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal azul/Am G30.txt'
bAmG40 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal azul/Am G40.txt'
bAmG50 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal azul/Am G50.txt'

bCsG30 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal azul/Cs G30.txt'
bCsG40 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal azul/Cs G40.txt'
bCsG50 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal azul/Cs G50.txt'

bCoG30 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal azul/Co G30.txt'
bCoG40 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal azul/Co G40.txt'
bCoG50 = '/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Raw_Data/Gamma Spectrocospy/cristal azul/Co G50.txt'

df_bAmG30 = pd.read_table(bAmG30, decimal=',')
df_bAmG40 = pd.read_table(bAmG40, decimal=',')
df_bAmG50 = pd.read_table(bAmG50, decimal=',')

df_bCsG30 = pd.read_table(bCsG30, decimal=',')
df_bCsG40 = pd.read_table(bCsG40, decimal=',')
df_bCsG50 = pd.read_table(bCsG50, decimal=',')

df_bCoG30 = pd.read_table(bCoG30, decimal=',')
df_bCoG40 = pd.read_table(bCoG40, decimal=',')
df_bCoG50 = pd.read_table(bCoG50, decimal=',')

df_bAm = [df_bAmG30, df_bAmG40, df_bAmG50]
df_bCs = [df_bCsG30, df_bCsG40, df_bCsG50]
df_bCo = [df_bCoG30, df_bCoG40, df_bCoG50]

for i in df_bAm:
    i.columns = ['ADC', 'Counts']

for i in df_bCs:
    i.columns = ['ADC', 'Counts']

for i in df_bCo:
    i.columns = ['ADC', 'Counts']
```

```python
for i, df in enumerate(df_bAm, start=1):
    plt.hist(df['ADC'],
             bins=500,
             weights=df['Counts'],
             alpha=0.7,
             log=True,
             label=f'Gain {10*i+20}')


plt.title('Blue Crystal Am')
plt.legend()
plt.grid()

plt.savefig('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Data_Processing/Blue Am Spectrum.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
for i, df in enumerate(df_bCs, start=1):
    plt.hist(df['ADC'],
             bins=500,
             weights=df['Counts'],
             alpha=0.7,
             log=True,
             label=f'Gain {10*i+20}')


plt.title('Blue Crystal Cs')
plt.legend()
plt.grid()

plt.savefig('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Data_Processing/Blue Cs Spectrum.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
for i, df in enumerate(df_bCo, start=1):
    plt.hist(df['ADC'],
             bins=250,
             weights=df['Counts'],
             alpha=0.7,
             log=True,
             label=f'Gain {10*i+20}')


plt.title('Blue Crystal Co')
plt.legend()
plt.grid()

plt.savefig('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Data_Processing/Blue Co Spectrum.png', dpi=300, bbox_inches='tight')

plt.show()
```
