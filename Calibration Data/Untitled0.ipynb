```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/Calibration Data/Histogram Checker.csv')

df
```

```python
df.describe()
```

```python
pd.plotting.scatter_matrix(df, figsize=(10,10))

plt.show()
```

```python
fig = plt.figure(figsize=(8,6), layout='constrained')

# scatter plot
sns.lineplot(data=df,
        x='% Error',
        y='Resolution %',
        hue = 'Crystal',
        palette = ['olive', 'blue', 'red'])

plt.xlabel('Resolution %', fontsize=18)
plt.ylabel('% Error', fontsize=18)
plt.xticks(fontsize=15)
plt.legend(title='Crystal', title_fontsize = 15, fontsize=15)

plt.savefig('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/New_Graphs/Resolution Error Relation.svg', dpi=300, bbox_inches='tight')

plt.show()
```

```python
fig = plt.figure(figsize=(8,6), layout='constrained')

sns.barplot(data=df,
            x = 'Crystal',
            y = '% Error',
            hue = 'Sample',
            palette='rocket'
)

plt.xlabel('Crystal', fontsize=18)
plt.ylabel('% Error', fontsize=18)
plt.xticks(fontsize=15)
plt.legend(fontsize=15)

plt.savefig('/content/drive/Shareddrives/Physics Lab 4/Gamma Ray Spectroscopy/New_Graphs/Percentage Errors.svg', dpi=300, bbox_inches='tight')

plt.show()
```
